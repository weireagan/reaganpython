```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = int(input ('n = '))
x = np.arange(0,n)
y = np.sin(x * np.pi / 180.0)
```

```python
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("sin.pic")
plt.xlim(-20,330)
plt.ylim(-1
         .5,1.5)
plt.show()
```
